## <font color='steelblue'>The Game </font>

```python
# introducing the game and identifying bugs that are not yet worked out. 

"""
DocString:
    
    A) Introduction:
    
    Welcome to The Legend of Xadiavatar, where spirits, dragons, elves, & humans inhabit the world. 
    This game is a trivia text adventure game that combines the worlds of the Avatar frinchise with The dragon Prince! 
    It will be especially fun for players familier with the franchises. 

    The purpose of this game is to make it to Republic City, the only city where humans and magical creatures live 
    together in peace and harmony.
    
    Part 1: The fountain of self-reflection
    Part 2: varies (depending on part 1 answers)
    Part 3: Katolis - lava of doom
    Part 4: Yue Bay - one last hurrah
    
    
    Links to Intros and Trailers
    - ATLA: https://www.youtube.com/watch?v=pJUgCzEdx9k
    - TLOK: https://www.youtube.com/watch?v=B3Mcq_XFgvI
    - TDP: https://www.youtube.com/watch?v=PWEtCsi3Eo8
    
    
    B) Known Bugs and/or Errors:
    None.

"""

# importing needed functions

#from sys import exit
#from random import randint


# _____________________________________________________________________________________________________

# starting the game with an introduction to the player as well as a while countdown loop
print(f"""
{'*' * 110}
""")
    
def start():
    global player_name
    global place

    print(f"""Welcome to The Legend of Xadiavatar, where spirits, \N{Ghost}, dragons \N{dragon}, elves \N{Elf}, & humans \N{Woman} share a world.
    
    {'*' * 110}
    
    You are on a journey to Republic City, the only city where humans and magical creatures live together in harmony
    A feast awaits you at Air Temple Island. Will you make it on time? 
    Follow along to find out by answering each question as best as you can.
    See you on the other side! \N{Slightly Smiling Face} 
    
     {'*' * 110}
    
    The countdown to your takeoff begins in...
    """)
    
    
    # declaring an object
    count = 5

    # looping while count is greater than zero
    while count > 0:
        print(count)
        count = count - 1
        input("< Press ENTER to continue. >\n")
    # liftoff
    print("your air ship just took off!")


    input('<Press enter to continue>\n')
    
    # prompting user for input on name as dynamic input to be used throughout the game
    # using casefold in all player inputs (dynamic and non dynamic) 
    # to account for different spellings of the answers to game questions
    
    # using capitalize to ensure that user name and place are capitalized wherever they appear in the game
    
    
    player_name = input(prompt = "What is your name?\n")
    player_name = player_name.casefold()
    player_name = player_name.capitalize()
    input('<Press enter to continue>\n')
    
    # prompting user for input on a place to be used in the game
    
    place = input(prompt = "Where is your least favourite place in the world? \n")
    place = place.casefold()
    place = place.capitalize()
    input('<Press enter to continue>\n')    
    self_reflection()
    
    
# _____________________________________________________________________________________________________

# creating self_reflection() as the first stage of the game 
# which allows players to choose a different second stage of the game each time 
# depending on ther answers in this stage

def self_reflection():
    global player_name
    global place
    print(f"""
    
    It's a brand new day and I hope you are not in {place}. 
    You are starting out your journey at the fountain of self-reflection. 
    Which of these words best describes your current mood?
    
        1. cool
        2. grounded
        3. calm
        4. fiery
    """)
    
    mood = input(">  \n")
    mood = mood.casefold()

# using an if/elif/else conditional to determine where the player goes next

    if "1" in mood or "cool" in mood or "one" in mood:
        Southern_Water_Tribe()
    
    elif "2" in mood or "grounded" in mood or "two" in mood:
        Omashu()
    
    elif "3" in mood or "calm" in mood or "three" in mood:
        Northern_Air_Temple()
        
    elif "4" in mood or "fiery" in mood or "four" in mood:
        Ba_Sing_Se()
        
    else: 
        print("Invalid command. Please try again.")
        input("<Press enter to try again>\n")
        self_reflection()


# _____________________________________________________________________________________________________

#defining the second stage of the game based on mood 1: Southern_Water_Tribe

def Southern_Water_Tribe():
    global player_name
    global place
    print(f""" 
    
    
    Very cool, {player_name} \N{winking face} , 
    you are now at the Southern Water Tribe, home to legendary waterbending \N{Water Wave} master, Katara.
    {'*' * 110}
    Answer this question to move on to the next stage of your journey""")
    
    
    pipsqueak = input(prompt ="""Who is the leader of the freedom fighters? Jet or Ski?""")
    pipsqueak = pipsqueak.casefold()
    
    #creating an if/elif/else conditional for player to move on to stage 3 of the game
    
    if pipsqueak == 'jet' or pipsqueak == '1' or pipsqueak == 'one':
        print("""You are correct!""")
        input("<Press enter to continue>\n")
        Katolis()
        
    elif pipsqueak == 'ski' or pipsqueak == '2' or pipsqueak == 'two':
        print ("""That is incorrect. Perhaps it's time for for some more self reflection.""")
        input("<Press enter to continue>\n")
        self_reflection()
        
    else:
        print ("Invalid Entry. Please try again.")
        Southern_Water_Tribe()
    
    
    
# _____________________________________________________________________________________________________

#defining the second stage of the game based on mood 2: Omashu

def Omashu():
    global player_name
    global place
    print(f""" 
    
    
    It's always nice to feel grounded, {player_name}. 
    You are now at Omashu, home to the legendary earthbending beavers!
    {'*' * 110}
    Answer this question to move on to the next stage of your journey""")
    
    
    beifong = input(prompt ="""Who is the blind bandit?
    1. Suki
    2. Toph
    """)
    beifong = beifong.casefold()
    
    #creating an if/elif/else conditional for player to move on to stage 3 of the game
    
    if beifong == 'suki' or beifong == '1' or beifong == 'one':
        print ("""That is incorrect. Perhaps it's time for for some more self reflection.""")
        input("<Press enter to continue>\n")
        self_reflection()
    
    elif beifong == 'toph' or beifong == '2' or beifong == 'two':
        print("""You are correct!""")
        input("<Press enter to continue>\n")
        Katolis()
            
    else:
        print ("Invalid Entry. Please try again.")
        Omashu()
    
    
    
# _____________________________________________________________________________________________________

#defining the second stage of the game based on mood 3: Northern_Air_Temple

def Northern_Air_Temple():
    global player_name
    global place
    print(f""" 
    
    
    Good thing you're feeling calm, {player_name}, because you have arrived at the Northern Air Temple, 
    home to the most peaceful monks in Xadiavatar. \N{Person in Lotus Position}
    {'*' * 110}
    Answer this question to move on to the next stage of your journey""")
    
    
    aang = input(prompt ="""Who was the avatar before Aang?
    1. Roku
    2. The Blue Bandit
    """)
    aang = aang.casefold()
    
    #creating an if/elif/else conditional for player to move on to stage 3 of the game
    
    if aang == 'roku' or aang == '1' or aang == 'one' or aang == 'avatar roku':
        print("""You are correct!""")
        input("<Press enter to continue>\n")
        Katolis()
    
    elif aang == 'the blue bandit' or aang == '2' or aang == 'two' or aang == 'blue bandit' or aang == 'bandit' or aang == 'blue':
        print ("""That is incorrect. Perhaps it's time for for some more self reflection.""")
        input("<Press enter to continue>\n")
        self_reflection()
    
    else:
        print ("Invalid Entry. Please try again.")
        Northern_Air_Temple()
    
    
    
# _____________________________________________________________________________________________________

#defining the second stage of the game based on mood 4: Ba Sing Se


def Ba_Sing_Se():
    global player_name
    global place
    print(f"""
   
    
    Welcome to Ba Sing Se, {player_name} ! It is time for test your knowledge on this great city, 
    so great thing you're feeling fired up.  
    {'*' * 110}
    Who led the six hundred-day Siege of Ba Sing Se? 
    (hint: this fire-nation legend loves tea \N{Teacup Without Handle} and Pai Sho, often together)""")
    print ("""
        1. The Cabbage Merchant 
        2. General Iroh
        3. Azula
        4. Bambi""")
            
    choice = input(">  \n")
    choice = choice.casefold()

# using an if/elif/else conditional with "in" to determine where the player goes next

    if "1" in choice or "the cabbage merchant" in choice or "one" in choice or "cabbage" in choice or "merchant" in choice:
        print("\nSorry, that is incorrect. Please try again.\n")
        input("<Press enter to continue>\n")
        Ba_Sing_Se()
    
    elif "2" in choice or "general iroh" in choice or "two" in choice or "iroh" in choice or "general" in choice:
        print("\nCorrect! You are off to Katolis.")
        input("<Press enter to continue>\n")
        Katolis()
    
    elif "3" in choice or "azula" in choice or "princess azula" in choice:
        print("\nShe wishes she did, but sadly she was a child during that battle\n")
        input("<Press enter to try again>\n")
        Ba_Sing_Se()
        
    elif "4" in choice or "bambi" in choice or "four" in choice:
        print("\n I'm afraid that is incorrect. \N{Slightly Frowning Face}")
        fail() #This moves us into fail
        
    else: 
        print("Invalid command. Please try again.")
        input("<Press enter to try again>\n")
        Ba_Sing_Se()

    
    
    
# _____________________________________________________________________________________________________

# defining third stage of the game: Katolis 


def Katolis():
    global place
    print(f"""
    
    
    Congratulations {player_name}, you are now in the human Kingdom of Katolis!
    {'*' * 110}
    You need to cross the lava of doom and only Sol Regem can carry you safely across. 
    {'*' * 110}
    For him to do this, you must answer the following question correctly""")
    
    
    dragon_type = input(prompt = """What type of dragon is Sol Regem?
    (hint: it's one of six primal elements: Sky, Earth, Ocean, Star, Sun or Moon) in:\n""")
    dragon_type = dragon_type.casefold()
    
    input('<Press enter to continue>\n')

    
    #using if/elif/else conditional to determine if player will move on to stage 4 or start this stage all over again

    
    if dragon_type == 'sun'or dragon_type =='sun dragon' or dragon_type == 'sun':
        print(f"""Correct! You are unstoppable 
              (but be careful since that's just another kind of stoppable, courtesey of Claudia \N{winking face} \N{Crystal Ball})\n
              """)
        input("<Press enter to continue>\n")
        Yue_Bay() # This moves us into Yue Bay
        
    elif dragon_type == 'star'or dragon_type == 'sky' or dragon_type == 'earth'or dragon_type == 'moon' or dragon_type == 'ocean':
        print('This answer is incorrect. Please try again')
        input('<Press enter to continue>\n')
        Katolis()
        
    elif dragon_type == '':
        fail() # This moves us into fail
    
    else:
        print("Invalid entry. Please try again.\n")
        input('<Press enter to try again>\n')
        Katolis() # Brings us back to the beginning of Katolis to try again

        
        
# _____________________________________________________________________________________________________        
        
# defining the fourth and final stage of the game, Yue_Bay

def Yue_Bay():
    global place
    print(f"""
    You have made it the shores of Yue Bay, just outside Republic City. I hope that dragon ride wasn't too tiresome. 
    {'*' * 110}
    You have one final question, {player_name}.
    """)
    
    sport = input(prompt ="""What is the most famous sport in Republic city? 
    (hint: it's either pro-bending or kyoshi fan)""")
    sport = sport.casefold()
    
    # using if/elif/else conditional to get player to Republic City
    
    if sport == 'pro-bending' or sport == 'pro bending' or sport == 'probending':
        print("""That is absolutely correct \N{Party Popper}""")
        input("<Press enter to continue>\n")
        win()
        
    elif sport == 'kyoshi fan' or sport == 'kyoshi' or sport == 'fan':
        print ('You are now on an ariship back to Katolis')
        input("Press enter to continue")
        Katolis()
        
    else:
        print ("Invalid Entry")
        fail()


# defining win and fail variables as the final stages of the game
# using drumroll emojis if the player wins the game

def win():
    print(f"""Drumroll please!
    
    
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941
    \U0001F941\U0001F941\U0001F941\U0001F941\U0001F941

    {'*' * 110}

    """)

    print (f"""
    Congratulations {player_name}, you have made it to Republic City! 
    {'*' * 110}
    Lin Beifong and Meelo are waiting to escort you to Air Temple Island""")
    input("<Press enter to exit!>\n")
    exit(0)

def fail():
    global place
    print(f""" You are now on a bus to {place}. Better luck next time.""")
    input('<Game Over \N{Slightly Frowning Face}>')
    question_1 = input (prompt ='Do you want to play again? yes/no')
    
    if question_1 == 'yes' or question_1 == 'y':
        print(f"""{'*' * 110}""")
        self_reflection()
        
    else:
        print(f"""I hope you come back soon!
        {'*' * 100}
        """)
        exit(0)

start()
```
